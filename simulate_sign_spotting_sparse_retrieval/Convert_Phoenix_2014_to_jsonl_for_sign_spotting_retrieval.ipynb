```python
import pandas as pd
from pathlib import Path
```

```python
df = pd.concat([pd.read_csv(csv_file, delimiter="|") for csv_file in Path.cwd().glob("*.csv")])
df
```

```python
df["doc_id"] = df["id"].str.split("_default-").str[0]
df
```

```python
# count nans in source_video
df["doc_id"].isna().sum()
```

```python
# for each unique source_video, concatenate all "annotation"
video_transcripts = []
for source_video, source_vid_df in df.groupby("doc_id"):
    print(source_video)
    concatenated = source_vid_df["annotation"].str.cat(sep=" ")
    video_transcripts.append({"source_video": source_video, "transcript": concatenated})
```

```python
# save as jsonl
import json
with open("video_transcripts.jsonl", "w") as f:
    for video_transcript in video_transcripts:
        f.write(json.dumps(video_transcript) + "\n")
```

```python
# for each unique source_video, concatenate all "annotation"
video_transcripts = []
for id, source_vid_df in df.groupby("id"):
    print(id)
    concatenated = source_vid_df["annotation"].str.cat(sep=" ")
    video_transcripts.append({"id": id, "transcript": concatenated})
```

```python
# save as jsonl
import json
with open("segment_transcripts.jsonl", "w") as f:
    for video_transcript in video_transcripts:
        f.write(json.dumps(video_transcript) + "\n")
```
